```python
def k1(a,b):
    return gcd(a,b)**2/(a*b)

def k2(a,b):
    return gcd(a,b)/(a+b)

def rad(n):
    return prod(prime_divisors(n))

def k3(a,b):
    return 1/sigma(a*b/gcd(a,b)**2)

def k4(a,b):
    return gcd(a,b)/max(a,b)

def k5(a,b):
    return gcd(a,b)**3/(a*b*(a+b))

def k6(a,b):
    return gcd(a,b)/(a+b-gcd(a,b))

def k7(a,b):
    return gcd(a,b)/min(a,b)
```
